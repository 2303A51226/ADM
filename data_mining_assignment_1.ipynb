loading of dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/2017_Taxi_Trip_Data.csv")
df.head()
df.info(verbose=True)
print(df.columns)

plt.show()



```

handling the missing values by mean median or mode imputation

```python
 print(df.isnull().sum())
```

```python
df['extra'].fillna(df['extra'].mean(), inplace=True)
```

```python
df['extra'].fillna(df['extra'].median(), inplace=True)
```

Min-Max Normalization

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

columns_to_normalize = ['extra', 'fare_amount']

df[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize])

df[columns_to_normalize].head()
```

Standardization

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize])

df[columns_to_normalize].head()
```
